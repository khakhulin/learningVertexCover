```python
from matplotlib import pyplot as plt
import numpy as np
import pickle as pkl
import matplotlib
```

```python
def average(values, num_bins=100):
    values = np.array_split(np.asarray(values), num_bins)
    values = [v.mean() for v in values]
    return np.asarray(values)
```

```python
for experiment_name in ['erdos_renyi', 'powerlaw', 'barabasi_albert', 'gnp_random_graph']:
    losses = pkl.load(open(f'logs/{experiment_name}_min-10_max-100.pkl', 'rb'))

    plt.figure(figsize=(10, 7))
    matplotlib.rcParams.update({'font.size': 22})
    plt.title('Reward')
    plt.xlabel('Epoch')
    plt.ylabel('| MVC |')
    plt.plot(-1 * average(losses['reward']))
    plt.savefig(f'{experiment_name}_reward.png')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 7))
    matplotlib.rcParams.update({'font.size': 22})
    plt.title('Value Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Value Loss')
    plt.plot(average(losses['value_loss']))
    plt.savefig(f'{experiment_name}_value_loss.png')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 7))
    matplotlib.rcParams.update({'font.size': 22})
    plt.title('Policy Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Policy Loss')
    plt.plot(-1 * average(losses['policy_loss']))
    plt.savefig(f'{experiment_name}_policy_loss.png')
    plt.tight_layout()
    plt.show()
```

```python
for experiment_name in ['erdos_renyi', 'powerlaw', 'barabasi_albert', 'gnp_random_graph']:
    heuristic = np.load(f'logs/{experiment_name}_min-10_max-100_heuristic.npy')
    agent = np.load(f'logs/{experiment_name}_min-10_max-100_agent.npy')

    plt.figure(figsize=(10, 7))
    matplotlib.rcParams.update({'font.size': 22})
    plt.title('Comaprison with heuristic')
    plt.xlabel('Graph size')
    plt.ylabel('| MVC |')
    plt.plot(heuristic.mean(1))
    plt.plot(agent.mean(1))
    plt.legend(['heuristic', 'agent'])
    plt.savefig(f'{experiment_name}_heuristic-comp.png')
    plt.show()
```
